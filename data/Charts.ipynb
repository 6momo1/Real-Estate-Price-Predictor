```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("./cleaned_data/cleaned_vancouver.csv",encoding = "ISO-8859-1")
```

```python
df.columns
```

```python
# a scatter plot comparing num_children and num_pets
df.plot(kind='scatter',x='lot_size',y='price',color='red')
plt.show()
```

```python

df = df.replace(np.nan,0)
series_price_vs_area = df.groupby(['Sub-Area/Community'])['price'].mean()


series_price_vs_area = series_price_vs_area.sort_values()
series_price_vs_area.plot.bar(x='Community', y='Price')
```
