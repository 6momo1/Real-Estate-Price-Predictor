```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("./cleaned_data/cleaned_vancouver.csv",encoding = "ISO-8859-1")
```

```python
df.columns
```

```python
# a scatter plot comparing num_children and num_pets
df.plot(kind='scatter',x='lot_size',y='price',color='red')
plt.show()
```

```python
df_sorted = df.sort_values('price')
df_sorted.plot(kind='bar',x='Sub-Area/Community',y='price')
df_sorted.groupby('Sub-Area/Community')['price'].nunique().plot(kind='bar')
plt.show()
```

```python
df_sorted = df.sort_values('price')
df.plot(kind='bar',x='Sub-Area/Community',y='price')
df.groupby('Sub-Area/Community')['price'].nunique().plot(kind='bar')
plt.show()
```
