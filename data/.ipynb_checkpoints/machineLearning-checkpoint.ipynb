```python
import pandas as pd

df = pd.read_csv("./cleaned data/pandas_vancouver.csv",encoding = "ISO-8859-1")
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.head()
```

```python
dummies = pd.get_dummies(df.area)
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['area','Renfrew'],axis=1)
final.columns
```
