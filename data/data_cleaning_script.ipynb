```python
import pandas as pd

df = pd.read_csv("./cleaning_data/data_vancouver.csv",encoding = "ISO-8859-1")
df.columns
df.head()
# remove useless columns (only the first 15 columns are useful)
```

```python
# drop the columns that contains Unamed
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.head(3)
```

```python
# return the columns that equal to Na
df[df['property_age'].isna()] 
```

```python
#describe
df.describe()
```

```python
# drop all rows that contain NA
df = df.dropna()
```

```python
# filter rows that are houses only
df = df[df['Property_Type'].str.contains("House")]
df.head()
```

```python
# This cell is used to filter the "lot_size" column

# filter1: data rows with lot size in the format: AREAft^2
# type: pandas.core.series.Series
filter1 = (~df['lot_size'].str.contains('(', regex=False))

# filter2: data rows with lot size in the format: WIDTH ft X LENGHT ft (AREAft^2)
filter2 = ~filter1

# to apply the filter1
df[filter1]

# show the index of each row
# type: pandas.core.indexes.numeric.Int64Index
# output: dtype='int64', length=456
filter2.index


# drop method rows with filter ( Doesn't work? )
# df.drop(filter2.index)

# alternate method using DF ( works )
df = df[filter2]
df.head()
```

```python
# this cell extracts the text inisde the brackets in the "lot_size" column
# and manipulates the data to return the digits only for the "lot_size"



type(df['lot_size']) # pandas.core.series.Series
type(df)             # pandas.core.frame.DataFrame

# convert the dtype to string
df['lot_size'].astype("string")


# function to extract text inside bracket
def get_value_inside_brackets(s):
    return s[s.find("(")+1:s.find(")")]


# apply the funciton
df['lot_size'] = df['lot_size'].apply(get_value_inside_brackets)


# function that returns the digits only in lot_size
def digit_only(string):
    # create a string that contains only digits
    string_digit = ''.join(char for char in string if char.isdigit())
    
    # special case: remove the last character in this string because of the ^2 sign
    string_digit = string_digit[0:-1]
    return int(string_digit)


df['lot_size'] = df['lot_size'].apply(digit_only)

df.head()
```

```python
# extract the age of a house from the "property_age" and convert it as dtype: int64


# convert the dtype to string
df['property_age'].astype("string")


# function to extract text inside bracket
def get_value_inside_brackets(s):
    return s[s.find("(")+1:s.find(")")]


# apply the changes
df['property_age'] = df['property_age'].apply(get_value_inside_brackets)


# function that returns the digits only in property_age
def digit_only_for_age(string):
    # create a string that contains only digits
    string_digit = ''.join(char for char in string if char.isdigit())
    
    return int(string_digit)


df['property_age'] = df['property_age'].apply(digit_only_for_age)

df.head()
```

```python
# convert the "price" values to dtype: int64 by removing commas and "$" sign


# convert the dtype to string
df['price'].astype("string")


# function to extract text inside bracket
def get_value_inside_brackets(s):
    return s[s.find("(")+1:s.find(")")]


# apply the changes
df['price'] = df['price'].apply(get_value_inside_brackets)


# function that returns the digits only in "price" column
def digit_only(string):
    # create a string that contains only digits
    string_digit = ''.join(char for char in string if char.isdigit())
    
    # return it as int type
    return int(string_digit)


df['price'] = df['price'].apply(digit_only)

df.head()
```

```python
# convert the "gross_tax_for_2020" values to dtype: int64 by removing "," and "$" sign


# convert the dtype to string
df['gross_tax_for_2020'].astype("string")


# function that returns the digits only in "price" column
def digit_only(string):
    # create a string that contains only digits
    string_digit = ''.join(char for char in string if char.isdigit())
    
    if string_digit == '':
        return 0
    
    # return it as int type
    return int(string_digit)


df['gross_tax_for_2020'] = df['gross_tax_for_2020'].apply(digit_only)

df.head()
```
