```python
import pandas as pd

df = pd.read_csv("./cleaned_data/cleaned_vancouver.csv",encoding = "ISO-8859-1")
df.head(3)
```

```python
# dummies variable 
dummies = pd.get_dummies(df['Sub-Area/Community'])
dummies.head(3)
```

```python

merged = pd.concat([df, dummies], axis='columns')
merged.head(3)
```

```python
# drop a column from the dummy variables to avoid the dummy variable trap
final = merged.drop(['Sub-Area/Community','Southlands'],axis='columns')
final.head(3)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# drop unecessary columns
final = final.drop(['address','area','Property_Type', 'title', 'style','heating_type', 'Depth', 'frontage'],axis='columns')
```

```python
X = final.drop('price', axis='columns')
X.head(3)
```

```python
y = final.price
y.head(3)
```

```python
# create a model
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
# one hot encoder method

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# transform Community column with label encoded method

dfle = df

# drop unnecessary columns
dfle = dfle.drop(['address','area','Property_Type', 'title', 'style','heating_type', 'Depth', 'frontage'],axis='columns')

# label encode Coummnity column
dfle['Sub-Area/Community'] = le.fit_transform(dfle['Sub-Area/Community'])
dfle.head()
```

```python
X = dfle.drop(['price'],axis='columns')
X = X.values
X
```

```python
y = dfle.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('Sub-Area/Community', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```
