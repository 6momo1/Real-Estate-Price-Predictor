```python
import pandas as pd

df = pd.read_csv("../cleaned_data/cleaned_vancouver.csv",encoding = "ISO-8859-1")
df.head(3)
```

```python
# dummies variable 
dummies = pd.get_dummies(df['Sub-Area/Community'])
dummies.head(3)
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged.head(3)
```

```python
# drop a column from the dummy variables to avoid the dummy variable trap
final = merged.drop(['Sub-Area/Community','Southlands'],axis='columns')
final.head(3)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# drop unecessary columns
final = final.drop(['address','area','Property_Type', 'title', 'style','heating_type', 'Depth', 'frontage'],axis='columns')
```

```python
X = final.drop('price', axis='columns')
X.head(3)
```

```python
y = final.price
y.head(3)
```

```python
# create a model
model.fit(X,y)
```

```python
model.score(X,y)
```
